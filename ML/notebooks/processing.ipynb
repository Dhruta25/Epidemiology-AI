```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Users/dhrutamacm2/Desktop/Epidemiology/ML/data/raw/Disease_symptom_and_patient_profile_dataset.csv")
```

```python
df = df.drop(columns=["Outcome Variable"])
```

```python
binary_cols = ["Fever", "Cough", "Fatigue", "Difficulty Breathing"]

df[binary_cols] = df[binary_cols].replace({
    "Yes": 1,
    "No": 0
})
```

```python
bp_map = {"Low": 0, "Normal": 1, "High": 2}
chol_map = {"Low": 0, "Normal": 1, "High": 2}

df["Blood Pressure"] = df["Blood Pressure"].map(bp_map)
df["Cholesterol Level"] = df["Cholesterol Level"].map(chol_map)
```

```python
df["Gender"] = df["Gender"].map({
    "Male": 1,
    "Female": 0
})
```

```python
x = df.drop(columns=['Disease'])
y = df['Disease']
```
